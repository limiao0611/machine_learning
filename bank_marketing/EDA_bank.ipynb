```python
# separate numeric variables and categorical variables
from pandas.api.types import is_numeric_dtype, is_string_dtype
num_cols = []
cat_cols = []
for col in train.columns:
    if is_numeric_dtype(train[col]):
        num_cols.append(col)
    elif is_string_dtype(train[col]):
        cat_cols.append(col)
        
print(num_cols)
cat_cols.remove('y')
print(cat_cols)
```

```python
# get a summary of y
train.y.value_counts()
```

```python
import numpy as np
import matplotlib.pylab as plt

plt.bar([1,2], train.y.value_counts())
plt.xticks([1,2], train.y.value_counts().index)
ratio = round(train.y.value_counts()[1]/train.y.size,3)
plt.title('y,  yes fraction: '+ str(ratio))
```

## imbalanced data, base rate = 0.117

# Univariate analysis:
## categorical variables -> bar chart
## numeric variables -> histogram

```python
plt.figure(figsize=(15,18))
for i,col in enumerate(num_cols):
    ax1 = plt.subplot(4,2,i+1 )
    ax2 = ax1.twinx()
   # ax1.set_xlabel(col)
    ax1.set_yscale('log')
    range_min = train[col].min()
    range_max = train[col].max()
    num0='(y=0 or 1)'
    if col == 'pdays':
        num0 = num0 + ', # of -1:'+str(train[train[col]==-1][col].size)
    elif col == 'previous':
        num0 = num0 + ', # of 0:'+str(train[train[col]==0][col].size)
    hist_val = ax1.hist(train[col], bins=30, label=col+num0, range=(range_min, range_max))
    hist_val1 = ax1.hist(train[train.y=='yes'][col], bins=30, label='y=1', range=(range_min, range_max))
    ax1.legend()
    
    ax2.plot([range_min, range_max], [ratio, ratio], c='black', ls='--', label='base rate of 1:'+str(ratio))
    ax2.plot(hist_val[1][1:], hist_val1[0]/hist_val[0], c='black', label='fraction of 1 in the bin' )
    ax2.set_ylabel('fraction of 1')
    if col == 'balance':
        ax2.legend(loc=5)
```

```python
plt.figure(figsize=(15, 40))
frac_devi = []
for i, col in enumerate(cat_cols):
    if i==0: 
        ax1 = plt.subplot(6, 2, (1,2))
    else:
        ax1 = plt.subplot(6, 2, i+2)
    #ax1 = train[col].value_counts().plot(kind='bar')
    value_counts = train[col].value_counts()
    ax1.bar(range(value_counts.size), value_counts, label='y=1 or 0' )
    ax1.set_xticks(range(value_counts.size), value_counts.index, rotation=40 )
    ax1.set_title(col, fontsize=15)
    
    value_counts1 = train[train.y=='yes'][col].value_counts()
    ax1.bar(range(value_counts1.size), value_counts1, label='y=1')    
    ax1.set_yscale('log')
    ax1.legend()
    
    ax2 = ax1.twinx()
    ax2.plot(range(value_counts.size), value_counts1/value_counts, c='black', marker='s', 
             label = 'fraction of 1 in the bin')
    ax2.plot([-1, value_counts.size], [ratio]*2, c='black', ls='--', label='base rate: '+str(ratio))
    ax2.set_ylabel('fraction of 1')
    
    frac_devi.append([value_counts.size, np.mean(value_counts1/value_counts), 
                      np.std(value_counts1/value_counts),
                      np.array(value_counts1/value_counts)
                     ])
    
    if col == 'job':
        ax2.legend(loc=5)
    
```

```python
frac_devi = np.array(frac_devi).T
frac_devi
```

```python
frac_devi_T = frac_devi.T
```

```python
plt.errorbar(range(len(frac_devi_T[1])), frac_devi_T[1], yerr = frac_devi_T[2], 
             label='mean +/- std base rate in each category')
for i, col in enumerate(cat_cols):
    plt.scatter(frac_devi[i][0]*[i], frac_devi[i][3])
plt.xticks(range(len(frac_devi_T[1])), cat_cols, rotation=30)
plt.plot([0, frac_devi_T[1].size], [ratio]*2, '--', lw=3, label='base rate')
plt.ylabel('y=1 fraction')
plt.legend()
plt.title('Categorical Variables')
```
